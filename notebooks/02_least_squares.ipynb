# Least Squares Estimation for SIR Parameters

In this notebook, we estimate the SIR model parameters using the least squares method.

We use synthetic observations of infected individuals \( I_{\text{obs}}(t) \), generated from the SIR model with noise.

The goal is to estimate:
- infection rate \( \beta \)
- recovery rate \( \gamma \)

## 1. Import libraries and set path

We import required libraries and add the project root to Python path
so that modules in `src` can be used.

```python
import sys
import os

sys.path.append(os.path.abspath("../"))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.model import solve_sir
from src.estimation import estimate_parameters_ls
```

## 2. Load synthetic data

We load the synthetic observations generated in the previous notebook.

```python
df = pd.read_csv("../data/synthetic_sir.csv")
df.head()
```

## 3. Define observed data and initial conditions

We assume that only the infected population is observed.

We also specify the initial conditions used in the SIR simulation.

```python
t = df["t"].values
I_obs = df["I_obs"].values

S0 = 999
I0 = 1
R0 = 0

beta_true = 0.3
gamma_true = 0.1
```

## 4. Least squares formulation

We estimate the parameters by minimizing the sum of squared errors:

\[
\sum_t \left(I_{\text{obs}}(t) - I_{\text{model}}(t; \beta, \gamma)\right)^2
\]

This compares the observed infected trajectory with the model-generated trajectory.

## 5. Run least squares estimation

We use numerical optimization to find the values of \( \beta \) and \( \gamma \)
that minimize the least squares objective.

```python
ls_result = estimate_parameters_ls(
    t=t,
    I_obs=I_obs,
    S0=S0,
    I0=I0,
    R0=R0,
    beta_init=0.2,
    gamma_init=0.2
)

ls_result
```

## 6. Extract estimated parameters

```python
beta_hat = ls_result["beta_hat"]
gamma_hat = ls_result["gamma_hat"]

print(f"Estimated beta  : {beta_hat:.4f}")
print(f"Estimated gamma : {gamma_hat:.4f}")
print(f"True beta       : {beta_true:.4f}")
print(f"True gamma      : {gamma_true:.4f}")
```

## 7. Simulate the fitted SIR model

Using the estimated parameters, we simulate the SIR model again
and compare the fitted trajectory with the observed data.

```python
S_fit, I_fit, R_fit = solve_sir(beta_hat, gamma_hat, S0, I0, R0, t)
S_true, I_true, R_true = solve_sir(beta_true, gamma_true, S0, I0, R0, t)
```

## 8. Compare true trajectory, observations, and fitted trajectory

```python
plt.figure(figsize=(9, 5))

plt.plot(t, I_true, label="True I(t)", linewidth=2)
plt.scatter(t, I_obs, label="Observed I(t)", s=18, alpha=0.7)
plt.plot(t, I_fit, label="Fitted I(t) by LS", linewidth=2, linestyle="--")

plt.xlabel("Time")
plt.ylabel("Infected")
plt.title("Least Squares Fit for SIR Model")
plt.legend()
plt.show()
```

## 9. Compare parameter values

```python
comparison_df = pd.DataFrame({
    "parameter": ["beta", "gamma"],
    "true_value": [beta_true, gamma_true],
    "estimated_value": [beta_hat, gamma_hat]
})

comparison_df
```

## 10. Interpretation

If the fitted trajectory is close to the observed data
and the estimated parameters are close to the true values,
then the least squares estimation works well for this toy example.

In practice, estimation quality depends on:
- noise level
- identifiability
- quality of initial guesses
- which compartments are observed
